```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('titanic.csv')
data
```

Часть 1: Разбиение исходной таблицы на две: по полу пассажиров

Таблица, соответствующая женскому полу

```python
female_data = data[data['Sex'] == 'female']
female_data
```

Таблица, соответствующая мужскому полу

```python
male_data = data[data['Sex'] == 'male']
male_data
```

Часть 2: Удаление пустых строк

```python
male_data = male_data.dropna()
female_data = female_data.dropna()
data = data.dropna()
```

Часть 3: Построение гистограммы распределения пассажиров по возрасту

```python
plt.figure(figsize=(10, 6))
plt.hist(data['Age'], bins=20, color='lightseagreen', edgecolor='black')

plt.title('Распределение пассажиров по возрасту', fontsize=16)
plt.xlabel('Возраст', fontsize=14)
plt.ylabel('Количество пассажиров', fontsize=14)

plt.show()
```

Часть 4: Вычисление мин и макс, среднего, моды, медианы возраста пассажира каждого пола

```python
group_by = data.groupby('Sex')

age_stats = pd.DataFrame({
    'Мин.возраст': group_by['Age'].min(),
    'Макс.возраст': group_by['Age'].max(),
    'Средний возраст': group_by['Age'].mean(),
    'Мода возраста': group_by['Age'].apply(lambda x: x.mode()[0]),
    'Медиана возраста':  group_by['Age'].median()
})

age_stats
```

Часть 5: Визуализация возраста при помощи различных настроек метода plot

```python
age_bins = pd.cut(data['Age'], bins=[0, 18, 35, 50, 80])
age_bins.value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), colors=['steelblue', 'lightseagreen', 'skyblue', 'darkturquoise'], title='Распределение пассажиров по возрастам')

plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
data.boxplot(column='Age', by='Sex', grid = False, patch_artist=True, boxprops=dict(facecolor='steelblue'))
plt.title('Ящик с усами возраста пассажиров по полу')
plt.suptitle('')
plt.ylabel('Возраст')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
male_data['Age'].plot(kind='kde', label='Мужчины', color='steelblue')
female_data['Age'].plot(kind='kde', label='Женщины', color='lightseagreen')
plt.title('График плотности возраста пассажиров по полу')
plt.xlabel('Возраст')
plt.ylabel('Плотность')
plt.legend()
plt.show()
```

```python
mean_age = data.groupby('Sex')['Age'].mean()
mean_age.plot(kind='bar', figsize=(8, 6), color=['steelblue', 'lightseagreen'], title='Средний возраст пассажиров по полу')

plt.ylabel('Средний возраст')
plt.show()
```
